```python
import pandas as pd
import sqlite3
from cleantext import clean

from pathlib import Path
import swifter
```

```python
conn = sqlite3.connect("/home/filter/ptf/data/million_post_corpus/corpus.sqlite3")
```

```python
comments = pd.read_sql('select headline, body, createdat from posts', conn)
```

```python
comments = comments.fillna('')
```

```python
comments['text'] = comments.swifter.apply(lambda x: clean(x['Headline'] + ' ' + x['Body'], lang='de'), axis=1)
```

```python
comments['month'] = comments['CreatedAt'].apply(lambda x: x[5:7])
```

```python
comments['year'] = comments['CreatedAt'].apply(lambda x: x[:4])
```

```python
c16 = comments.query('year=="2016"')
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]
```

```python
c16 = c16[c16['text'].str.len() > 50]
```

```python
c16['month'].value_counts()
```

```python
out = '/home/filter/ptf/dynamic_bernoulli_embeddings/dat/arxiv_ML/raw'
```

```python
c16.query('month=="' + '01' + '"')
```

```python
for x in ['01', '02', '03', '04', '05']:
    li = c16.query('month=="' + x + '"')['text'].values.tolist()
    print(len(li))
    for i, c in enumerate(chunks(li, 1000)):
#         print(c)
        out_str = '\n'.join(c)
        Path(out + '/' + x + '_' + str(i) + '.txt').write_text(out_str)
```

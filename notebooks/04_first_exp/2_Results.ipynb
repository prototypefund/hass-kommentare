```python
from pathlib import Path
import pandas as pd
```

```python
arr = []
for r in Path('/mnt/data2/ptf/co_exp/lemma_nostop_hyper/res2').glob('*'):
    try:
        txt = r.read_text().split()[-1]
        arr.append({'res': float(txt), 'ds':r.name.split('.')[0], 'params':' '.join(r.name.split('.')[1:])})
    except:
        pass
# sorted(arr, key=lambda x: x[0])
```

```python
df = pd.DataFrame(arr)
```

```python
df.groupby('params').sum().reset_index().sort_values('res', ascending=False).head(5).merge(df, on='params')
```
